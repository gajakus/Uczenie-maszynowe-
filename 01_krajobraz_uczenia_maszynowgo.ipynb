```python
import matplotlib.pyplot as plt 
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from pathlib import Path

np.random.seed(42)

# domyslne rozmiary czcionek 
plt.rc('font', size=12)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize =12)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

pd.set_option("display.max_columns", None)

```

```python
datapath = Path() /"data" / "lifesat" #definiujemy ścieżke 
# wczytujemy zbiory
oecd_bli = pd.read_csv(datapath / "oecd_bli.csv")
gdp_per_capita = pd.read_csv(datapath / "gdp_per_capita.csv")
```

```python
oecd_bli
```

```python
gdp_per_capita
```

```python
#wstepnie przetwarzamy dane PKB per capita, tak by uwzgledniany byl tylko 2020 rok i upraszamy nazwy kolumn
gdp_per_capita_2020 = gdp_per_capita[gdp_per_capita["Year"]==2020]
gdp_per_capita_2020 = gdp_per_capita_2020.rename(
    columns={
        "GDP per capita, PPP (constant 2017 international $)"
        : "GDP per capita (USD)"
    }
)
gdp_per_capita_2020
```

```python
oecd_bli["Indicator"].value_counts()
```

```python
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(
    index= "Country",
    columns="Indicator",
    values="Value"
)
oecd_bli
```

```python
oecd_bli = oecd_bli.reset_index()
oecd_bli
```

```python
gdp_per_capita_2020.info()
```

```python
oecd_bli.info()
```

```python
full_country_stats = oecd_bli.merge(
    gdp_per_capita_2020,
    left_on="Country",
    right_on="Entity",
    how="inner"
)
full_country_stats
```

```python
full_country_stats = full_country_stats[[
    "Country", "Life satisfaction", "GDP per capita (USD)"
]]
full_country_stats = full_country_stats.sort_values(
    by="GDP per capita (USD)"
)
full_country_stats
```

```python
# wyeksportujemy pelny zbior lifesat_full do pliku
full_country_stats.to_csv(
    datapath / "lifesat_full.csv",
    index=False
)
```

```python
min_gdp = 23_500 #łatwiej sie to czyta uzytkownikowi, Python to pomija
max_gdp = 62_500

country_stats  = full_country_stats[
    (full_country_stats["GDP per capita (USD)"] >= min_gdp)
    & (full_country_stats["GDP per capita (USD)"] <= max_gdp)
]
country_stats.to_csv(
    datapath / "lifesat.csv",
    index = False
)
country_stats
```

```python
lifesat = pd.read_csv("data/lifesat/lifesat.csv")

#przygotowujemy dane
lifesat.rename(
    columns={"GDP per capita (USD)": "PKB na mieszkanca (USD)"},
    inplace=True
)
lifesat.rename(columns={"Life satisfaction": "Satysfakcja z życia"},
               inplace=True)
lifesat
```

```python
#Wizualizacja danych
lifesat.plot(kind='scatter', grid=True,
             x="PKB na mieszkanca (USD)", y='Satysfakcja z życia')
plt.show()
```

```python
# przygotowuje zmienne w modelu
X = lifesat[["PKB na mieszkanca (USD)"]]
y = lifesat[["Satysfakcja z życia"]]

#wybiera model liniowy
model = LinearRegression()

#trenuje model (algorytm MNK)
model.fit(X, y)

#obliczamy współczynniki modelu
theta_0=model.intercept_[0]
theta_1=model.coef_[0][0]
print(f"theta_0: {theta_0}")
print(f"theta_1: {theta_1}")

```

```python
#  wizualizujemy dane
lifesat.plot(kind='scatter', grid=True,
             x="PKB na mieszkanca (USD)", y='Satysfakcja z życia')
plt.plot(X, theta_0+theta_1*X)
plt.text(45000, 6.5, fr"$\theta_0 = {theta_0*round(2)}$",
          color= "blue")
plt.text(45000, 6.3, fr"$\theta_1 = {theta_1* 1e5:.2f}\cdot 10^{{-5}}$", #:.2 czyli zaokrągla do 2 miejsc po przecinku
          color= "blue")
plt.show()
```

```python
cyprus_gdp_per_capita_2020 = (
    gdp_per_capita_2020[gdp_per_capita_2020["Entity"]=="Cyprus"][["GDP per capita (USD)"]]
)
cyprus_gdp_per_capita_2020.rename(
    columns={"GDP per capita (USD)": "PKB na mieszkanca (USD)"},
    inplace=True
)
cyprus_gdp_per_capita_2020
```

```python
cyprus_predicted_life_satisfaction = model.predict(cyprus_gdp_per_capita_2020)
cyprus_predicted_life_satisfaction
```

```python
#biera model regresji 3 najlizszych sąsiadów
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor(n_neighbors=3)

#trenuje model_knn
model_knn.fit(X,y)
# uzysuje przeiwdwania dla cypru
model_knn.predict(cyprus_gdp_per_capita_2020)
```
